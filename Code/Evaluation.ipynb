<h3>deBERTaV3 base finetuned on SQuaD v2 finetuned AmazonQA:</h3>

```python
from haystack.nodes import FARMReader

reader_finetuned = FARMReader("./model_finetuned_amazonqa", top_k=10, return_no_answer=False, use_gpu=True)
```

```python
# Evaluate Reader on its own
reader_eval_results_finetuned = reader_finetuned.eval_on_file("./","./test_split_squad_format.json")
```

```python
top_n = reader_eval_results_finetuned["top_n"]

# Reader Top-N-Accuracy is the proportion of predicted answers that match with their corresponding correct answer including no_answers
print(f"Reader Top-{top_n}-Accuracy:", reader_eval_results_finetuned["top_n_accuracy"])
# Reader Top-1-Exact Match is the proportion of questions where the first predicted answer is exactly the same as the correct answer including no_answers
print("Reader Top-1-Exact Match:", reader_eval_results_finetuned["EM"])
# Reader Top-1-F1-Score is the average overlap between the first predicted answers and the correct answers including no_answers
print("Reader Top-1-F1-Score:", reader_eval_results_finetuned["f1"])
# Reader Top-N-Accuracy is the proportion of predicted answers that match with their corresponding correct answer excluding no_answers
print(f"Reader Top-{top_n}-Accuracy (without no_answers):", reader_eval_results_finetuned["top_n_accuracy_text_answer"])
# Reader Top-N-Exact Match is the proportion of questions where the predicted answer within the first n results is exactly the same as the correct answer excluding no_answers (no_answers are always present within top n).
print(f"Reader Top-{top_n}-Exact Match (without no_answers):", reader_eval_results_finetuned["top_n_EM_text_answer"])
# Reader Top-N-F1-Score is the average overlap between the top n predicted answers and the correct answers excluding no_answers (no_answers are always present within top n).
print(f"Reader Top-{top_n}-F1-Score (without no_answers):", reader_eval_results_finetuned["top_n_f1_text_answer"])
```

```python
import json
```

```python
with open('deberta_v3_base_squad2_amazonqa_evaldata.json', 'w') as f:
    json.dump(reader_eval_results_finetuned, f)
```

```python
del reader_finetuned, reader_eval_results_finetuned, top_n, f
```
