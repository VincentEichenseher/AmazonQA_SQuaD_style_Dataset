```python
import logging

logging.basicConfig(format="%(levelname)s - %(name)s -  %(message)s", level=logging.WARNING)
logging.getLogger("haystack").setLevel(logging.INFO)
```

```python
from haystack.nodes import FARMReader
from pathlib import Path
```

```python
base_model = "deepset/deberta-v3-base-squad2"
```

```python
reader = FARMReader(model_name_or_path=base_model, use_gpu=True, top_k=10, batch_size=8, return_no_answer=False)
```

```python
reader.train(data_dir="./",
             train_filename="train_split_squad_format.json",
             test_filename="test_split_squad_format.json",
             dev_filename="val_split_squad_format.json",
             use_gpu=True,
             batch_size=8,
             n_epochs=5,
             learning_rate=1e-5,
             evaluate_every=2500,
             max_seq_len=400,
             checkpoint_root_dir=Path("model_checkpoints"),
             checkpoint_every=5000,
             checkpoints_to_keep=3,
             save_dir="model_finetuned_amazonqa")
```
